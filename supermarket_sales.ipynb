```python
df['Date']=pd.to_datetime(df['Date'],format='%m/%d/%Y')
```

```python
df['Time']=pd.to_datetime(df['Time'],format='%H:%M').dt.time
```

```python
df['DateTime']=pd.to_datetime(df['Date'].astype(str)+' '+df['Time'].astype(str),format='%Y-%m-%d %H:%M:%S')
```

```python
x=['Branch','City','Customer type','Product line','Payment']
```

```python
for i in x:
    plt.figure(figsize=(12,5))
    sns.countplot(x=df[i])
    plt.title(f'Bar plot of {i}',color='blue')
    plt.xlabel(i)
    plt.ylabel('Count')
    plt.show()
```

```python
for i in x:
    df[i].value_counts().plot.pie(autopct='%1.1f%%')
    plt.title(f'Pie chart of {i}')
    plt.ylabel('')
    plt.show()
```

```python
for i in x:
    print(df[i].value_counts())
    print()
```

```python
y=['Unit price','Tax 5%','Total','cogs','gross income','Rating']
```

```python
for i in y:
    print(i)
    print('---------')
    print(f'mean: {df[i].mean()}')
    print(f'median: {df[i].median()}')
    print(f'min_val: {df[i].min()}')
    print(f'max_val: {df[i].max()}')
    print(f'variance: {df[i].var()}')
    print(f'standard_deviation: {df[i].std()}')
    print()
    print()
```

```python
for i in y:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    min_range=q1-(1.5*iqr)
    max_range=q3+(1.5*iqr)
    print(f'{i} | q1: {q1}, q3: {q3}, iqr: {iqr}, min_range: {min_range}, max_range: {max_range}')
    print()
```

```python
for i in y:
    plt.figure(figsize=(12,5))
    plt.boxplot(df[i])
    plt.title(f'Box plot of {i}',color='blue',size=16)
    plt.show()
```

```python
for i in y:
    plt.figure(figsize=(10,4))
    plt.hist(df[i],bins=10)
    plt.title(f'Histogram of {i}')
    plt.xlabel(i)
    plt.ylabel('frequency')
    plt.show()
```

```python
for i in y:
    sns.violinplot(y=df[i], color="lightblue")
    plt.title(f"Violin Plot of {i}")
    plt.ylabel(i)
    plt.show()
```

```python
corr_matrix=df[y].corr()
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
for i in y:
    print(f"Count of {i} by Branch:")
    print(df.groupby('Branch')[i].count())
    print(f"Sum of {i} by Branch:")
    print(df.groupby('Branch')[i].sum())
    print()
    print(f"Mean of {i} by Branch:")
    print(df.groupby('Branch')[i].mean())
    print()
    print(f"Median of {i} by Branch:")
    print(df.groupby('Branch')[i].median())
    print()
    print(f"Minimum of {i} by Branch:")
    print(df.groupby('Branch')[i].min())
    print()
    print(f"Maximum of {i} by Branch:")
    print(df.groupby('Branch')[i].max())
    print()
    print(f"Variance of {i} by Branch:")
    print(df.groupby('Branch')[i].var())
    print()
    print(f"Standard deviation of {i} by Branch:")
    print(df.groupby('Branch')[i].std())
    print('.....................')
```
